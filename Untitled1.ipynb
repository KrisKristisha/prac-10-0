# **Решение примера из текста**

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import os

def nash_equilibrium(ar):

    from fractions import Fraction
    import scipy
    from scipy.optimize import linprog

    print("Input: ar =\n", ar, "\n")
    v1 = np.array([1 for x in range(len(ar[0]))])
    v2 = np.array([1 for x in range(len(ar))])
    print("v1 = ", v1)
    print("v2 = ", v2, "\n")
    # cоздаю единичные векторы v1 и v2 длины первой строки и первого столбца матрицы

    limits = [(0, None)]
    for x in range(len(ar[0]) - 1):
        limits += [(0, None)]

    ar_T = -np.transpose(ar)
    res = linprog(v1, A_ub = ar_T, b_ub = -v2, bounds = limits, options = {"disp": False})
    p = [res.x[i]/(res.fun) for i in range(len(res.x))]

    #build_graf
    fig = plt.figure(figsize=(7, 5))  # добавила прямоугольник для рисования
    plt.title("Optimal strategy for the first player")
    for i in range(len(p)):
        scatter1 = plt.scatter(i+1, p[i])
    grid1 = plt.grid(True)   # линии сетки

    print ('Optimal value:', Fraction(1/(res.fun)).limit_denominator(1000))
    print('| P', end=" ")
    for i in range(len(p)):
        print ('|', Fraction(p[i]).limit_denominator(1000), end=" "),

    res = linprog(-v1, A_ub = ar, b_ub = v2, bounds = limits, options = {"disp": False})
    q = [res.x[i]/(-res.fun) for i in range(len(res.x))]

    #build_graf
    fig = plt.figure(figsize=(7, 5))  # добавила прямоугольник для рисования
    plt.title("Optimal strategy for the second player")
    for i in range(len(q)):
        scatter1 = plt.scatter(i+1, q[i])
    grid1 = plt.grid(True)   # линии сетки

    print ('|\n| Q', end=" ")
    for i in range(len(q)):
        print ('|', Fraction(q[i]).limit_denominator(1000), end=" "),
    print ('|')
    return

plt.show()

if __name__ == "__main__":
    import numpy as np
    ar = np.array([[ 4, 0, 6, 2, 2, 1],
                   [ 3, 8, 4, 10, 4, 4],
                   [ 1, 2, 6, 5, 0, 0],
                   [ 6, 6, 4, 4, 10, 3],
                   [ 10, 4, 6, 4, 0, 9],
                   [ 10, 7, 0, 7, 9, 8]])
    nash_equilibrium(ar)
    print ('\nИгра, в которой спектр оптимальной стратегии состоит из одной точки (т.е. существует равновесие Нэша в чистых стратегиях)')
```
